```python
# variance , std
temp = [29, 31, 33, 33, 32, 31, 30]

variance = np.var(temp)
print(variance)

std = np.std(temp)
print(std)
```

```python
# percentile , iqr, z-score
wt = [2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]
q1 = np.percentile(wt, 25)
q2 = np.percentile(wt, 50)
q3 = np.percentile(wt, 75)

iqr = q3-q1

print(q1)
print(q2)
print(q3)
print(iqr)

# outlier detection
lower_outlier = q1 - 1.5 * iqr
upper_outlier = q1 + 1.5 * iqr

print(lower_outlier)
print(upper_outlier)
```

```python
# z-score

z_scores = (65 - 50)/10
print(z_scores)

x = 50+((-2.5)*10)
print(x)
```

```python
# conditional probability and bayes
# Bayes’ Theorem
# Formula: P(A | B) = [P(B | A) × P(A)] / P(B)

# P(S∣F) = The probability the email is spam after we know it contains “free.”
# P(F∣S) = The probability an email contains “free” given that it is spam.  = 40/100
# P(S) = The probability an email is spam before we know anything else.    = 100/1000
# P(F) = The probability an email contains “free,” whether spam or not.    = 40+60/1000

total_email = 1000

P_S = 100/1000
P_F = 100/1000
P_F_given_S = 40/100
P_S_given_F = (P_F_given_S * P_S)/P_F
print(f"{P_S_given_F:.2f}")

# If an email contains the word “free”, there is a 40% chance it is spam.
```

```python
# base rate effect , answer done in khata using confusion matrix
# bayes theorem is

P_D = 1/10000
p_plus_D = 0.99
p_plus = 0.99+100
p_D_plus = (p_plus_D * P_D)/p_plus
print(p_D_plus)


# accuracy is 98%
# Even though the test is 99% accurate, the disease is extremely rare (1 in 10,000).
# base-rate fallacy is We tend to ignore how rare the disease is and focus only on the test accuracy.
```

```python
# Accuracy = (TP + TN) / Total
# Precision = TP / (TP + FP)
# Recall = TP / (TP + FN)
# F1 = 2 × Precision × Recall / (Precision + Recall)
# Specificity = TN / (TN + FP)
# NPV = TN / (TN + FN)
# Prevalence = (Actual Positive) / Total


accuracy = (120+770)/1000
print(accuracy)
# 89% of all predictions are correct.
# Good overall performance but can be misleading if classes are imbalanced.

precision = 120/(120+80)
print(precision)
# Of all predicted positives, 60% were actually positive.
# Lower precision → many false positives.

recall = 120/(120+30)
print(recall)
# The model catches 80% of actual positives.
# Good for detecting positives, but misses 20%.

f1 = 2*precision*recall/(precision+recall)
print(f1)
# Balance between precision and recall.
# F1 ≈ 0.686 → moderate performance.

Specificity = 770/(770+80)
print(Specificity)
# Correctly identifies 90.6% of actual negatives.
# Very good at avoiding false alarms.

npv = 770/(770+30)
print(npv)
# If the model predicts negative, 96% chance it's truly negative.
# Indicates strong confidence in negative classifications.

Prevalence = (120+30)/1000
print(Prevalence)
# Only 15% of the data are positive cases.
# Low prevalence explains why NPV is high and precision is lower.
```
