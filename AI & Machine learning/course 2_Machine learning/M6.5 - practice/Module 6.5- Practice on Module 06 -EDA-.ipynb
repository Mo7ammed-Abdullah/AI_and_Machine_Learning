## Section 9 — Numeric Features versus Target (Boxplots)

Compare numeric feature distributions across different target classes.


```python
# TODO:
# For each numeric feature:
# 1. Plot a boxplot with the feature on y axis and target on x axis
# 2. Compare distributions for each target class

plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_col,1):
  plt.subplot(2,6,i)
  sns.boxplot(x=df[target_col], y = df[col])
  plt.title(col)

plt.tight_layout()
plt.show()
```

### Student Notes and Observations

_Which numeric features show clear differences between target classes?_

Large differences in distributions between target classes often signal strong predictive potential. If the distributions are almost identical, that feature may be less useful on its own.

## Section 10 — Pairplot and Correlation Analysis

Explore relationships between features and correlations.


```python
# TODO:
# 1. Create a pairplot with selected numeric features plus the target
# 2. Generate a correlation matrix for numeric features and target
# 3. Visualize the matrix with a heatmap
# 4. Sort correlations with the target column

sns.pairplot(df[['account_age_days','total_transactions_user','avg_amount_user','amount','is_fraud']], hue='is_fraud',diag_kind = 'hist')
plt.show
```

```python
corr_matrix = df[numeric_col + [target_col]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm",fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap for Numerical Features")
plt.tight_layout()
plt.show()
```

## Section 11 — Categorical Features versus Target (Crosstab and Stacked Bar Plot)

Examine how target proportions change across categories.


```python
# TODO:
# For each categorical feature:
# 1. Create a crosstab between the feature and target with normalize="index"
# 2. Plot a stacked bar chart of the proportions
# 3. Comment on how the target distribution changes across categories

for c in categorical_col:
  ct = pd.crosstab(df[c], df[target_col], normalize="index")
  print(f"\nProportion of fraud within {c}")
  display(ct)

  ct.plot(kind="bar", stacked=True, figsize=(6,4))
  plt.title(f"fraud proportion by {c}")
  plt.ylabel("proportion")
  plt.xticks(rotation=0)
  plt.tight_layout()
  plt.show()
```

### Student Notes and Observations

_Which categories seem more associated with a particular target class?_
